```python
#   TransactionID, isFraud     numerical 
num_cols = [col for col in train_df.columns if col not in cat_cols and col not in ["TransactionID", "isFraud"]]
```

```python
#     

def cat_Ratio(dataframe, col_name):
    print(pd.DataFrame({col_name: dataframe[col_name].value_counts(),
                        "Ratio": 100 * dataframe[col_name].value_counts() / len(dataframe)}))
    print("\n")
```

```python
for col in cat_cols:
    cat_Ratio(train_df, col)
```

```python
def num_Ratio(dataframe, numerical_col):
    print(dataframe[numerical_col].describe())

    print("\n")
```

```python
for col in num_cols:
    num_Ratio(train_df, col)
```

```python
for col in num_cols: 
    print(f"{col}: {len(train_df[col].value_counts())}")
```
